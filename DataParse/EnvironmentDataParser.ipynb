```python
import pandas as pd
df = pd.read_csv('C:/Users/Luyao Chen/OneDrive/Big Data Programming/Final/environment.csv')
df.head(5)
```

```python
df.dtypes
```

```python
def WeatherTypeParser(code):
    code = str(code)
    if code == 'nan':
        w = 'normal'
    elif 'RA'in code or 'DZ'in code or 'UP' in code :
        w = 'rain'
    elif 'SN'in code or 'SG'in code or 'IC'in code or 'PL' in code or 'GR' in code or 'GS' in code:
        w = 'snow'
    elif 'BR'in code or 'FG'in code or 'FU'in code or 'SA' in code :
        w = 'fog'
    else:
        w = 'normal'
    return w

df1 = df.loc[:,['DATE','HOURLYPRSENTWEATHERTYPE','HOURLYDRYBULBTEMPC','HOURLYWindSpeed']]
df1['HOURLYPRSENTWEATHERTYPE']= df1['HOURLYPRSENTWEATHERTYPE'].map(lambda x:WeatherTypeParser(x))

df2 = df1.rename(columns={"HOURLYPRSENTWEATHERTYPE": "WEATHER", "HOURLYDRYBULBTEMPC": "TEMPC","HOURLYWindSpeed":"WIND"})
df2.head(5)
```

```python
df2.head(5)
def WeatherDateParser(date):
     return date[0:-3]

df2['DATE']=df2['DATE'].map(lambda x:WeatherDateParser(x))
df2.head(5)
```

```python
df2.shape
df2.head(5)
```

```python
repeat = []
for i in range(0,df2.shape[0]-1):
    if df2.loc[i]['DATE'] == df2.loc[i+1]['DATE']:
        repeat.append(i)
```

```python
df2.drop(df.index[repeat], inplace=True)
df2.shape
```
